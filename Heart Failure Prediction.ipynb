```python
del df['Category']
```

```python
y = df['Mortality']
print(y)
```

```python
x = df.drop('Mortality', axis=1)
```

```python
x
```

```python
x.shape
```

```python
print(x.columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
for col in ['Age.Group', 'Gender', 'Locality  ', 'Marital status                       ', 'Life.Style                                                                              ', 'Sleep', 'Depression', 'Hyperlipi', 'Smoking', 'Family.History', 'HTN', 'Allergies', 'Hypersensitivity', 'Others ', 'CO', 'Diagnosis', 'SK.React' ]:
    x[col] = label_encoder.fit_transform(x[col])
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transform =  ColumnTransformer([("norm1",OneHotEncoder(),[0])], remainder = 'passthrough')
x = transform.fit_transform(x)
```

```python
x
```

#Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

#Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred_naive_bayes = classifier.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_naive_bayes))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_naive_bayes))
print("Classification Report:\n", classification_report(y_test, y_pred_naive_bayes))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_naive_bayes), annot=True)
```

#Decision Tree

```python

from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred_decission_tree = classifier.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_decission_tree))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_decission_tree))
print("Classification Report:\n", classification_report(y_test, y_pred_decission_tree))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_decission_tree), annot=True)
```

#K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5,
                                  metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
y_pred_knn = classifier.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_knn), annot=True)
```

#ANN

```python
x_train_all, x_test1, y_train_all, y_test1 = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train1, x_valid, y_train1, y_valid = train_test_split(x_train_all, y_train_all, test_size=0.15, random_state=42)
```

```python
from tensorflow import keras
model = keras.models.Sequential()

model.add(keras.layers.Dense(300, activation="sigmoid"))
model.add(keras.layers.Dense(100, activation="sigmoid"))
model.add(keras.layers.Dense(10, activation="softmax"))

# model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

```python
model_history = model.fit(x_train1,y_train1,epochs=15,validation_data=(x_valid,y_valid))
```

```python
model.evaluate(x_test1, y_test1)
```

```python
from sklearn.metrics import confusion_matrix

y_pred_ann = model.predict(x_test1)
y_pred_ann = np.argmax(y_pred_ann, axis=1)


conf_matrix_ann = confusion_matrix(y_test1, y_pred_ann)


plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_ann, annot=True, fmt="d", cmap="Greens", cbar=False)

plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
